### PCA Visualization

```python
import torch
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import torch
import torch.nn.functional as F

# Charger les features et labels
features = torch.load("output/abstract_features.pt")
labels = torch.load("output/abstract_labels.pt")



# Appliquer PCA pour rduire les features  2 dimensions
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(features)

# Afficher la visualisation
plt.figure(figsize=(10, 8))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=labels, cmap='viridis', s=20)
plt.colorbar()  # pour afficher la lgende de la couleur (en fonction des classes)
plt.title("2D PCA Visualization of features")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```
